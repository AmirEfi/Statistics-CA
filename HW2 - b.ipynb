```python
from scipy.stats import geom
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The number of maximum of freekicks for the question
x = np.arange(1,4)
# Probability of being goal of a freekick
prob = geom.pmf(x , p = 0.25)
print('Probability of being goal of a freekick before 4 tries: ', sum(prob))

plt.figure(figsize=(16,8))
plt.vlines(x, 0, prob, lw=15)
plt.xlabel("Freekicks", fontsize = 20)
plt.xticks(np.arange(1, 3+1, 1))
plt.ylabel("Probability of Goal", fontsize = 20)
```
